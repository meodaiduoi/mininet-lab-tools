```python
import pandas as pd
import os
```

```python

def ls_subfolders(rootdir):
    sub_folders_n_files = []
    for path, _, files in os.walk(rootdir):
        for name in files:
            sub_folders_n_files.append(os.path.join(path, name))
    return sorted(sub_folders_n_files)

for filename in ls_subfolders('../ds/'):
    try:
        print(filename)
        pd.read_csv(filename).sort_values(by=['time_epoch']).to_csv(filename, index=False)
    except Exception as e:
        print(filename, e)
```
